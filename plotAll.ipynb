```python
corr_grid = covid_data.corr()

groupings = ['date',
'critical_staffing_shortage',
'hospital_onset_covid',
'inpatient_beds',
'previous_day_admission',
['staffed_adult_icu_patients', 'staffed_icu_adult_patients'],
'total_adult_patients',
'total_pediatric_patients',
'total_staffed_adult_icu_beds',
'inpatient_beds_utilization',
'percent_of_inpatients_with_covid',
'inpatient_bed_covid_utilization',
'adult_icu_bed_covid_utilization',
'adult_icu_bed_utilization',
'previous_day_admission_adult_covid_confirmed',
'previous_day_admission_adult_covid_suspected',
'deaths_covid',
'on_hand_supply_therapeutic',
'icu_patients_confirmed_influenza',
'previous_day_admission_influenza_confirmed',
'previous_day_deaths_covid_and_influenza',
'previous_day_deaths_influenza',
'total_patients_hospitalized_confirmed_influenza',
'total_patients_hospitalized_confirmed_influenza_and_covid',
'all_pediatric_inpatient_bed_occupied',
'all_pediatric_inpatient_beds',
'previous_day_admission_pediatric_covid_confirmed',
'staffed_icu_pediatric_patients_confirmed',
'staffed_pediatric_icu_bed_occupancy',
'total_staffed_pediatric_icu_beds']

def in_same_group(h1, h2):
    for g in groupings:
        if type(g) is list:
            a = False
            b = False
            for sg in g:
                if h1.startswith(sg):
                    a = True
                if h2.startswith(sg):
                    b = True
            if a and b:
                return True
        elif h1.startswith(g) and h2.startswith(g):
            return True
    return False
            

pairs = []
def unique_pair(i, j):
    if i <= j:
        p = (i, j)
    else:
        p = (j, i)
    if p in pairs:
        return False
    pairs.append(p)
    return True

# https://stackoverflow.com/questions/34902649/how-do-i-loop-across-a-correlation-matrix-to-only-give-me-pairs-of-correlations
count = 0
for i in list(corr_grid.columns):
    for j in list(corr_grid.columns):
        if not in_same_group(i, j) and unique_pair(i, j):
            if abs(corr_grid.loc[i,j]) > 0.7 and corr_grid.loc[i,j] != 1:
                count += 1
                print(i, ' ',j ,' ', corr_grid.loc[i,j])
print(f"Total significant correlations: {count}")
```
